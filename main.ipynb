```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Configuración para que los gráficos se vean mejor
sns.set_theme(style="whitegrid")

# 1. Lista de archivos (usando los nombres exactos de tus archivos)
archivos = [
    'LaLiga16 17.xlsx - Worksheet.csv', 
    'LaLiga17 18.xlsx - Worksheet.csv', 
    'LaLiga18 19.xlsx - Worksheet.csv', 
    'LaLiga19 20.xlsx - Worksheet.csv', 
    'LaLiga20 21.xlsx - Worksheet.csv', 
    'LaLiga21 22.xlsx - Worksheet.csv'
]

temporadas = ['2016-17', '2017-18', '2018-19', '2019-20', '2020-21', '2021-22']

# 2. Carga y unificación en un solo DataFrame
lista_df = []

for archivo, temp in zip(archivos, temporadas):
    # Cargamos el CSV
    df_temp = pd.read_csv(archivo)
    
    # Agregamos la columna de temporada para no perder la referencia temporal
    df_temp['Temporada'] = temp
    
    # Limpieza inmediata de filas vacías (separadores de jornada en el Excel original)
    df_temp = df_temp.dropna(subset=['Local', 'Score', 'Visitante'])
    
    lista_df.append(df_temp)

# Unificamos todo
df = pd.concat(lista_df, ignore_index=True)

# 3. Procesamiento de Goles y Resultados
# Separamos el Score (ej: "2–1") en columnas numéricas
# Nota: A veces el guion es un carácter especial '–', lo manejamos con regex
df[['Goles_Local', 'Goles_Visitante']] = df['Score'].str.split('–|--', expand=True).astype(int)

#Creamos la variable de resultado para el equipo Local
def obtener_resultado(row):
    if row['Goles_Local'] > row['Goles_Visitante']:
        return 'Gana Local'
    elif row['Goles_Local'] < row['Goles_Visitante']:
        return 'Gana Visita'
    else:
        return 'Empate'

df['Resultado'] = df.apply(obtener_resultado, axis=1)

# Verificamos la carga
print(f"Total de partidos cargados: {len(df)}")
print(df.info())
df.head()
```
